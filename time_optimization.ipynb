```python
import numpy as np
import pandas as pd

def fast_groupby(df_cars):
    indices = df_cars['int_model']
    min_values = np.maximum.reduceat(df_cars['target_reg'].values[np.argsort(indices)], np.concatenate(([0], np.cumsum(np.bincount(indices))))[:-1]) 
    sum_values = np.bincount(df_cars['int_model'], weights=df_cars['target_reg'])
    count_values = np.bincount(df_cars['int_model'])
    mean_values = np.bincount(df_cars['int_model'], weights=df_cars['target_reg']) / count_values
    
    res = pd.DataFrame({'int_model': np.arange(len(min_values)),
                        'min': min_values,
                        'sum': sum_values,
                        'count': count_values,
                        'mean': mean_values})
    return res
```
