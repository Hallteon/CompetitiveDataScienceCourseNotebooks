```python
import matplotlib.pyplot as plt

def distance_sensor():
    # Функция имитирует показания датчика расстояния
    distance = 10  # Изначальное расстояние до стенки
    while distance > 0:
        distance -= 1  # Имитация движения платформы в направлении стенки
        yield distance

def control_signal(distance):
    # Функция управляет сигналом на электродвигателе платформы
    signal = 0
    while distance > 0:
        if distance > 5:
            signal = 1  # Устанавливаем сигнал для движения вперед
        elif distance <= 5:
            signal = -1  # Устанавливаем сигнал для движения назад
        yield signal

distances = distance_sensor()  # Генератор показаний датчика расстояния
signals = control_signal(next(distances))  # Генератор управляющих сигналов

distance_values = []
signal_values = []

while True:
    try:
        distance = next(distances)  # Получаем текущее расстояние от датчика
        signal = next(signals)  # Получаем текущий управляющий сигнал
        distance_values.append(distance)
        signal_values.append(signal)
        print(f"Расстояние: {distance}, Сигнал: {signal}")

        if distance <= 0:
            break  # Остановка платформы при достижении стены
        
        signals = control_signal(distance)  # Обновляем генератор управляющих сигналов
    except StopIteration:
        break

# Построение графика зависимости сигнала от расстояния
plt.plot(distance_values, signal_values)
plt.xlabel('Расстояние до стенки')
plt.ylabel('Сигнал управления скоростью')
plt.title('График зависимости сигнала от расстояния')
plt.show()
```
